# Project Overview : Life Expectancy and GDP

In this project, I am going to investigate whether there is a relationship between the GDP and life expectancy of six countries below:
- United States of America
- Mexico
- Zimbabwe
- Germany
- China
- Chile



This project will be using the following: Matplotlib, Seaborn, Pandas, and mostly basic Python concepts.

# Step 1 - Loading & Importing Info

Before we can make our data visualizations, we need to import the following:
- Matplotlib
- Seaborn
- Pandas
- all_data.csv file

```python
# Getting data and importing the necessary modules
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

country_data = pd.read_csv("all_data.csv")
country_data.head()
```

- From first glance, the data seems just fine until we notice that there is one column that is pretty long(Life Expectancy at birth(years))!
That needs to be fixed by renaming this column to something that is much simpler like life_expectancy

```python
# Rename column to make it more understandable to readers
country_data.rename(columns={"Life expectancy at birth (years)": "life_expectancy"}, inplace=True)
country_data.head()
```

Now that is fixed, we can finally start solving these questions for our project:
- Has life expectancy increased over time in the six nations?
- Has GDP increased over time in the six nations?
- Is there a correlation between GDP and life expectancy of a country?
- What is the average life expectancy in these nations?
- What is the distribution of that life expectancy?

# Step 2: Life Expectancy in the Six Nations 

To see how Life Expectancy changed over time in the six nations, we must create a line plot that shows all six nations below:

```python
# Call sns.relplot to create a relational plot
sns.set_theme(style="darkgrid")
# palette='colorblind' allows people who have colorblindness to see the graph clearly!
sns.lineplot(x = country_data.Year, y= country_data.life_expectancy, hue=country_data.Country, palette='colorblind')

plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0) #Moves the legend to outside
plt.show() # This code displays the plot
```

# Findings & Conclusions from the Plot

From this relational plot, it definitely seems like all countries have increased life expectancy, the biggest noticeable increase being Zimbabwe.
- In the year 2000, life expectancy in Zimbabwe was around 46-47 and went as low as 43 before increasing to slightly above 60 years in 2015(that is a 20-25 years in life-expectancy increase, amazing!)
- Germany tends to have the highest life expectancy around 81 years just above Chile



# Step 3 - GDP in the Six Nations

To showcase any significant changes in GDP for all six nations, we will use a relational plot again to display this data below:

```python
# Call sns.relplot to create a relational plot
sns.set_theme(style="darkgrid")
sns.relplot(x = country_data.Year, y= country_data.GDP, hue=country_data.Country, palette='summer')
plt.title("GDP over time")
plt.ylabel("GDP (in trillions)")
plt.show() # This code displays the plot
```

# Findings & Conclusions from the Plot (Step 3)

From the plot above, it seems like the following countries increased in GDP overtime: USA, China, Zimbabwe and Chile.
While some have somewhat decreased: Germany & Mexico

- The USA and China had the biggest noticeable increases in GDP, especially China since they were below Germany in GDP from 2000 to 2005!
- Germany has increased GDP from 2000 to 2007 before staying stagnant for the next few years until slowly decreasing in the 2010s.


# Step 4 - Correlation between GDP & Life Expectancy

- To figure out a relationship between GDP and Life Expectancy, we need to create a scatter plot to see any noticeable changes between the six countries below:


```python
ax = sns.scatterplot(data=country_data, y='GDP', x='life_expectancy', hue='Country', palette='deep')
ax.set_ylabel('GDP (in trillions)')
ax.set_xlabel('Life Expectancy (in years)')
ax.set_title('Life Expectancy vs GDP')
```

Since this data is not accurately portraying the differences, we will instead separate all six countries with their own data
to make it easier to read

```python
s = sns.FacetGrid(country_data, col='Country', hue='Country', col_wrap=3, sharey=False, sharex=False)
s.map(sns.scatterplot,'life_expectancy','GDP', palette='dark')
s.set_axis_labels('Life Expectancy (in years)','GDP (in trillons)')
s.fig.suptitle('Relationship of Life Expectancy and GDP', fontsize=14)
s.fig.tight_layout(rect=[0, 0, 1.0, 1.0])
```

# Findings & Conclusions from this Multi-plot graphs


There is a strong positive correlation for all six countries between life expectancy and GDP
- Chile's relationship with life expectancy and GDP is quite stunning with GDP going from just below one trillion to nearly 3 trillion before dropping down with life expectancy over 80 years.
- The most scattered data out there is definitely Zimbabwe and Germany


# Step 5 - Average & Distribution Life Expectancy in Six Nations

- To see the average life expectancy, we can create a boxplot to display this info:

```python
avg_life = sns.boxplot(data=country_data, y='Country', x='life_expectancy', palette='spring')
avg_life.set_xlabel("Life Expectancy (in years)")
avg_life.set_title("Life Expectancy Distribution")
plt.show()
```

# Averages for all Six Nations from Boxplot

- Zimbabwe by far has the lowest life expectancy average at around 47-48 years, whereas the highest average is from Germany with around 80 years(a 33-year difference!)
- The distribution for life expectancy seems to be spread out over 70 years for almost all of them except Zimbabwe which only had distributions from 45 - to slightly above 60.
